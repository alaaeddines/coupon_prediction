# Merging of tables

```python
dfCT['coupon_used']= dfCT.coupon_discount.apply(lambda x:1 if x !=0 else 0)
```

```python
dfCT['count']=1
```

```python
dfCT['year']=dfCT['date'].apply(lambda x: x.year)
dfCT['month']=dfCT['date'].apply(lambda x: x.month)
dfCT['day']=dfCT['date'].apply(lambda x: x.day)
dfCT['dow']=dfCT['date'].apply(lambda x: x.dayofweek)
```

```python
dfCT.head()
```

```python
dfCT1= pd.pivot_table(dfCT, index = 'item_id',
               values = ['customer_id','quantity','selling_price',
                         'other_discount','coupon_discount','count','coupon_used'],
               aggfunc={'customer_id':lambda x:len(set(x)),
                        'quantity':np.mean,
                        'selling_price':np.mean,
                        'other_discount':np.mean,
                        'coupon_discount':np.mean,
                        'count':sum,
                        'coupon_used':sum
                       }
              )
```

```python
new_columns = {'customer_id':'ct_cust_it',
            'quantity':'avg_order_quantity',
            'selling_price':'avg_price',
            'other_discount':'avg_oth_discount',
            'coupon_discount':'avg_cp_discount',
            'count':'ct_orders',
           'coupon_used':'sum_cp_used'}
```

```python
dfCT1.rename(columns=new_columns, inplace = True)
```

```python
dfCT1.info()
```

```python
dfCT1.head()
```

```python
dfCT1['sum_cp_discount'] = dfCT1['avg_cp_discount']*dfCT1['ct_orders']
```

```python
dfCT1['sum_oth_discount'] = dfCT1['avg_oth_discount']*dfCT1['ct_orders']
```

```python
dfCT1['sum_order_quantity']= dfCT1['avg_order_quantity']*dfCT1['ct_orders']
```

```python
dfCT1['sum_total_discount'] = dfCT1['sum_cp_discount'] +dfCT1['sum_oth_discount']
dfCT1['avg_total_discount'] = dfCT1['avg_cp_discount'] +dfCT1['avg_oth_discount']
```

Merge Transactions with Items

```python
dfCT1.info()
```

```python

dfItem.info()
```

```python
dfCT1_Item=pd.merge(dfCT1, dfItem, on = 'item_id', how ='left')
```

```python
dfCT1_Item.head()
```

Merge dfCT1_Item with coupons table

```python
dfCT1_Item_Cp=pd.merge(dfCoupon,dfCT1_Item, on= 'item_id', how = 'left')
```

```python
dfCT1_Item_Cp.head(20)
```

```python
dfCT1_Item_Cp2 = pd.pivot_table(dfCT1_Item_Cp, index ='coupon_id',
                                values=['item_id', 'coupon_id', 'brand', 'brand_type', 'category', 'ct_orders',
                                        'avg_cp_discount', 'sum_cp_used', 'ct_cust_it', 'avg_oth_discount', 'avg_order_quantity',
                                       'avg_price', 'sum_cp_discount', 'sum_oth_discount', 'sum_order_quantity',
                                       'sum_total_discount', 'avg_total_discount'],
                               aggfunc={
                                   'item_id': lambda x: len(set(x)),
                                   'brand': lambda x: len(set(x)),
                                   'brand_type': lambda x: x.unique()[0] if (len(x.unique())==1) else 'Both',
                                   'category':lambda x: x.mode()[0],
                                   'ct_orders': np.sum,
                                   'avg_cp_discount':np.mean,
                                   'sum_cp_used':np.sum,
                                   'ct_cust_it':np.sum,
                                   'avg_oth_discount':np.mean,
                                   'avg_order_quantity':np.mean,
                                   'ct_orders':np.sum,
                                   'avg_price':np.mean,
                                   'sum_cp_discount':np.sum,
                                   'sum_oth_discount':np.sum,
                                   'sum_order_quantity':np.sum,
                                   'sum_total_discount':np.sum,
                                   'avg_total_discount':np.mean
                               }
                              )

new_columns= {
                   'item_id': 'ct_items',#item count
                   'brand': 'ct_brands',
                   'category':'mode_cat'
    
                                  }
dfCT1_Item_Cp2.rename(columns = new_columns, inplace = True)
```

```python
dfCT1_Item_Cp2['ct_cat']= pd.pivot_table(dfCT1_Item_Cp, 
                                         index='coupon_id',
                                         values ='category',
                                         aggfunc =lambda x: len(set(x)))
```

```python
dfCT1_Item_Cp2.head()
```

merge dfCT1_Item_Cp2 and train

```python
dfCT1_Item_Cp2_train = pd.merge(dfTrain, dfCT1_Item_Cp2, on = 'coupon_id', how= 'left')
dfCT1_Item_Cp2_train.head()
```

Customer data

```python
dfCT_trans = pd.pivot_table(dfCT, index = 'customer_id',
                           values =['item_id', 'quantity', 'selling_price',
                                    'other_discount','coupon_discount','count','coupon_used'],
                           aggfunc={'item_id':len, 
                                    'quantity':np.mean,
                                    'selling_price':np.mean,
                                    'other_discount':np.mean,
                                    'coupon_discount':np.mean,
                                   'coupon_used':np.mean})
new_columns={'item_id':'ct_items_cust',
          'quantity':'avg_quantity_cust',
          'selling_price':'avg_price_cust',
          'other_discount':'avg_oth_discount_cust',
          'coupon_discount':'avg_cp_discount_cust',
          'coupon_used':'avg_cp_used_cust'}

dfCT_trans.rename(columns=new_columns, inplace= True)
```

```python
 dfCT_trans['sum_quantity_cust']= dfCT_trans['ct_items_cust']* dfCT_trans['avg_quantity_cust']
 dfCT_trans['sum_price_cust']= dfCT_trans['ct_items_cust']* dfCT_trans['avg_price_cust']
 dfCT_trans['sum_oth_disccount_cust']= dfCT_trans['ct_items_cust']* dfCT_trans['avg_oth_discount_cust']
 dfCT_trans['sum_cp_discount_cust']= dfCT_trans['ct_items_cust']* dfCT_trans['avg_cp_discount_cust']
 dfCT_trans['ct_cp_used_cust']= dfCT_trans['ct_items_cust']* dfCT_trans['avg_cp_used_cust']
```

```python
dfCT_trans.head()
```

```python
dfCT_trans = pd.merge(dfCT_trans, dfDemo, on ='customer_id', how ='left')
```

```python
dfCT_trans.head()
```

final merge

```python
dfCT1_Item_Cp2_train_ct = pd.merge(dfCT1_Item_Cp2_train, dfCT_trans, on='customer_id', how='left')
```
